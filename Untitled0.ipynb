```python
#işlem yapabilmek için eklendi!

# CSV dosyalarını okuma
c_data = pd.read_csv(c_file_path)
h_data = pd.read_csv(h_file_path)

# Gerekli sütunları belirleme
c_columns_to_keep = ['Location Name', 'City', 'State', 'Wheelchair Accessible',
                     'Education (ex: personal growth, k-12 supports, reading research, etc.)',
                     'Digital Literacy', 'Borough']
h_columns_to_keep = ['Borough', 'AirTemp', 'Day', 'Hour']

# Gerekli sütunları seçme
c_data = c_data[c_columns_to_keep]
h_data = h_data[h_columns_to_keep]

# Rastgele 500 satır alma
c_sample = c_data.sample(n=500, random_state=1)
h_sample = h_data.sample(n=500, random_state=1)

# Veri setlerinin boyutunu kontrol etme
print("c veri seti (örnek) boyutu: ", c_sample.shape)
print("h veri seti (örnek) boyutu: ", h_sample.shape)
```

```python
# Ortak sütun olarak "Borough" kullanarak birleştirme işlemi
merged_data = pd.merge(c_sample, h_sample, on='Borough', how='inner')

# Sonuçları görüntüleme
print("Birleştirilmiş veri seti boyutu: ", merged_data.shape)
print("Birleştirilmiş veri seti örnek veriler:")
print(merged_data.head())
```

```python
## Sıcaklık ve Wheelchair Accessible sütunları arasındaki ilişkiyi Nedir?

plt.figure(figsize=(14, 8))
sns.violinplot(x='Wheelchair Accessible', y='AirTemp', data=merged_data)
plt.title('Temperature Distribution by Wheelchair Accessibility')
plt.xlabel('Wheelchair Accessible')
plt.ylabel('Air Temperature')
plt.show()
```

```python
#KORELASYON

# Wheelchair Accessible sütununu sayısal forma dönüştürme
merged_data['Wheelchair Accessible Numeric'] = merged_data['Wheelchair Accessible'].map({'Yes': 1, 'No': 0})

# Korelasyon matrisi oluşturma
correlation_matrix = merged_data[['AirTemp', 'Wheelchair Accessible Numeric']].corr()

# Korelasyon matrisini görüntüleme
print("Korelasyon Matrisi:")
print(correlation_matrix)

# Korelasyon matrisini görselleştirme
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
#1: Mükemmel pozitif korelasyon. Bir değişken artarken diğeri de artar.
#0.5 ile 1: Güçlü pozitif korelasyon. Bir değişken artarken diğeri de genellikle artar.
#0 ile 0.5: Zayıf pozitif korelasyon. Bir değişken artarken diğeri de genellikle artar, ancak ilişki güçlü değildir.
#0: Hiçbir korelasyon yok. Değişkenler arasında doğrusal bir ilişki yoktur.
#0 ile -0.5: Zayıf negatif korelasyon. Bir değişken artarken diğeri genellikle azalır, ancak ilişki güçlü değildir.
#-0.5 ile -1: Güçlü negatif korelasyon. Bir değişken artarken diğeri genellikle azalır.
#-1: Mükemmel negatif korelasyon. Bir değişken artarken diğeri azalır.
```

```python
#Hangi şehirlerdeki bilgisayar merkezleri daha fazla tekerlekli sandalye erişilebilirliğine sahip?

# Şehir bazında toplam ve erişilebilir merkez sayısını hesaplama
city_accessibility = merged_data.groupby(['City', 'Wheelchair Accessible']).size().unstack(fill_value=0)
city_accessibility['Total'] = city_accessibility.sum(axis=1)

# Erişilebilirlik yüzdesi hesaplama
city_accessibility['Accessible_Percentage'] = (city_accessibility['Yes'] / city_accessibility['Total']) * 100

# Sonuçları görüntüleme
print(city_accessibility)
```

```python
# Şehir bazında erişilebilirlik yüzdesini görselleştirme
plt.figure(figsize=(14, 8))
sns.barplot(x=city_accessibility.index, y=city_accessibility['Accessible_Percentage'], palette='viridis')
plt.title('City-wise Percentage of Wheelchair Accessible Centers')
plt.xlabel('City')
plt.ylabel('Percentage of Accessible Centers')
plt.xticks(rotation=90)
plt.show()
```

```python
# Şehir bazında toplam ve erişilebilir merkez sayısını görselleştirme
city_accessibility_melted = city_accessibility[['Yes', 'No', 'Partially']].reset_index().melt(id_vars='City', value_vars=['Yes', 'No', 'Partially'], var_name='Accessibility', value_name='Count')

plt.figure(figsize=(14, 8))
sns.barplot(x='City', y='Count', hue='Accessibility', data=city_accessibility_melted, palette='viridis')
plt.title('City-wise Distribution of Wheelchair Accessible Centers')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
